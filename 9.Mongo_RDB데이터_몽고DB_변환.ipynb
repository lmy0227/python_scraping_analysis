```python
import pandas as pd
import pymysql
import sqlalchemy
import pymongo

#pymysql과 sqlalchemy를 연동
pymysql.install_as_MySQLdb()
from sqlalchemy import create_engine

engine= create_engine('mysql+mysqldb://python:python@localhost:3306/python_db',encoding='utf-8')
print(engine)

#Engine을 사용해서 DB에 연결
con = engine.connect()
member_df=pd.read_sql_table('member',con)

#DB 연결 종료
con.close()
member_df.head()
```

```python
import json

with open('data/member.json','r',encoding='utf-8')as file:
    contents = file.read()
    member_json=json.loads(contents)
```

```python
import pymongo

con = pymongo.MongoClient()
db = con.member_db
col = db.members
print(col)
```

```python
col.insert_many(member_json)
print(col.count_documents({}))
```

```python
#정당이 열린민주당인 국회의원은?
for member in col.find({'정당':'열린민주당'},{'_id':0,'이름':1,'나이':1,'정당':1}):
    print(member)
```

```python
#정당별 개수
for member in col.aggregate([
    {
        '$group':{
            '_id':'$정당',
            '정당개수':{'$sum':1}
        }
    },
        {
            '$project':{
                '정당':'$_id',
                '_id':0,
                '정당개수':1
            }
        }
    
]):
    print(member)
```

```python
#취미가 등산 인 국회의원은?
for member in col.find({'취미, 특기':{'$regex':'등산'}},{'_id':0,'이름':1,'나이':1,'정당':1,'취미, 특기':1}):
    print(member)
```

```python
#$정당이 열린민주당 이면서 비례대표인 국회의원은?
for member in col.find({'정당':'열린민주당','선거구':'비례'}):
    print(member)
```
