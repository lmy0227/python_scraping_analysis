#### 2. 모든 날짜를 행에 만들기
* 확진자가 없는 날의 데이터도 만들기

```python
data['확진일자'].value_counts()
```

```python
data.head(2)
```

```python
#맨끝에 있는 행을 선택(최초의 확진일자)
first_day = data.iloc[-1,7]
first_day
```

```python
# 맨위에 있는 행을 선택(최근의 확진일자)
last_day = data.iloc[0,7]
last_day
```

```python
#pd.date_range() 함수를 통해서 시작하는 날짜부터 끝나는 날짜까지 (그 사이의 모든 날짜)
#DatetimeIndex 객체를 생성
days = pd.date_range(first_day, last_day)
print(type(days))
days[:5]
```

```python
#DateTimeIndex 객체인 days를 DataFrame으로 생성
df_days = pd.DataFrame({'확진일자':days})
print(df_days.shape)
df_days.head()
```

```python
#확진일자별로 빈도수 구하기
daily_case=data['확진일자'].value_counts()
daily_case.head()
```

```python
#확진일자별로 빈도수를 구한 Series 객체를 DataFrame으로 변환
df_daily_case = daily_case.to_frame()
print(df_daily_case.shape)
daily_case.head()
```

```python
#컬럼명 변경 확진일자 -> 확진자수
df_daily_case.columns = ['확진자수']
df_daily_case.head(2)
```

```python
df_days.head(2)
```

```python
#확진자가 없는 날도 일자에 표현이 되도록 전체일자와 확진일자 데이터를 merge()로 합친다.
all_day = df_days.merge(df_daily_case, left_on='확진일자',right_on=df_daily_case.index,how='left')
print(all_day.shape)
all_day.tail()
```

```python
#확진자수의 결측치(outlier) 값은 0으로 채우고 값을 누적해서 더해준다.
#fillna() 함수는 결측치 값을 다른 값으로 채워줄 때 사용하는 함수
#누적확진 컬럼을 추가하고 누적된 값을 할당
#cumsum() 누적해주는 함수
all_day['누적확진'] = all_day['확진자수'].fillna(0).cumsum()
all_day.tail()
```

```python
all_day.head()
```

```python
#확진일자에서 연도는 제외하고 월-일로 '일자'컬럼 만들기
all_day['일자'] = all_day['확진일자'].astype(str).map(lambda x:x[-5:])
all_day.head()
```

```python
#확진자수 누적확진, 일자 컬럼을 선택해서 새로운 DataFrame을 생성하기
cum_day = all_day[['일자','확진자수','누적확진']]
#일자 컬럼을 index로 설정
cum_day = cum_day.set_index('일자')
cum_day
```

```python
# DataFrame으로 확진자수와 누적확진을 line plot으로 그리기
cum_day.plot(figsize=(15,4))
```

```python
#Series로 2개의 line plot을 그리기
cum_day['확진자수'].plot()
cum_day['누적확진'].plot(figsize=(15,4))
```

#### 확진월과 요일 구하기

```python
all_day.head(1)
```

```python
#확진월
all_day['확진월'] = all_day['확진일자'].dt.month
#확진요일
all_day['확진요일'] = all_day['확진일자'].dt.dayofweek
all_day.tail(2)
```

```python
#월별, 요일별 확진자수를 groupby로 구하기
all_day_week = all_day.groupby(['확진월','확진요일'])['확진자수'].sum()
all_day_week
```

```python
#unstack() 은 인덱스를 컬럼으로 변경해주는 함수
#fillna() 은 결측치(NaN)을 0으로 채워주는 함수
#astype(int)는 int타입으로 변환해주는 함수
all_day_week = all_day_week.unstack().fillna(0).astype(int)
all_day_week
```

```python
dayofweek = '월 화 수 목 금 토 일'
dayofweek = dayofweek.split()
print(type(dayofweek))
dayofweek
```

```python
#컬럼의 이름을 한글요일명으로 변경해줍니다.
all_day_week.columns = dayofweek
all_day_week
```
