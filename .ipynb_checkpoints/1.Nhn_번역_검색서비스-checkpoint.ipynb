파이썬 기본 urllib 라이브러리 사용

```python
# 네이버 Papago NMT API 예제
import os
import sys
import urllib.request

client_id = "6rQvKe23YCefmfVHYYfq"
client_secret = "uPYrjLwtFO"
encText = urllib.parse.quote("Yesterday all my troubles seemed so far away.")
data = "source=en&target=ko&text=" + encText
url = "https://openapi.naver.com/v1/papago/n2mt"
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request, data=data.encode("utf-8"))
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

requests 라이브러리를 사용

```python
import requests

client_id = "6rQvKe23YCefmfVHYYfq"
client_secret = "uPYrjLwtFO"
encText = "Yesterday all my troubles seemed so far away."
url = "https://openapi.naver.com/v1/papago/n2mt"

#request header 값 선언
req_header = {
    "X-Naver-Client-Id":client_id,
    "X-Naver-Client-Secret":client_secret
}
#request form data 값 선언
req_data ={
    "source":"en",
    "target":"ko",
    "text":encText
}

#request 보내기 .post()함수 호출
res=requests.post(url, data=req_data, headers=req_header)
print(type(res))
print(res.status_code, res.ok)
print('응답헤더',res.headers)
print('요청헤더',res.request.headers)
```

```python
if res.ok:
    print(type(res.text))
    print('응답 Text', res.text)
else:
    print('Status Code :', res.status_code)
```

```python
#원하는 값만 뽑아내기 위해 text가 아니라 json사용
if res.ok:
    result_json= res.json()
    print(type(result_json))
    print(result_json)
```

### papago 서비스 사용
* yesterday, txt 가사 파일의 내용을 읽어서 list에 저장
* list에 있는 내용을 번역 서비스 요청
* 서비스 응답 결과를 다시 파일에 쓰기

```python
import requests

def translyric_savefiles(trans_list):
    with open('data/yesterday_trans.txt','w',encoding='utf8')as file:
        file.writelines(trans_list)

#yesterday.txt 가사 파일의 내용 읽어서 list에 저장
def gettext_savelist():
    lyric_list =[]
    with open('data/yesterday.txt','r',encoding='utf8') as file:
        contents = file.read()
        lyric_list = contents.split('\n')
    return lyric_list

def main():
    client_id = "KadYYszcWbbF7ypJoRpu"
    client_secret = "qEPp5fNh4n"
    
    url = "https://openapi.naver.com/v1/papago/n2mt"

    #request header 값 선언
    req_header = {
        "X-Naver-Client-Id":client_id,
        "X-Naver-Client-Secret":client_secret
    }
    
    #yesterday.txt 가사 파일의 내용 읽어서 list에 저장
    lyric_list = gettext_savelist()
    print(len(lyric_list))
    lyric_list = [lyric for lyric in lyric_list if len(lyric) != 0]
    print(len(lyric_list))
    
    trans_list = []
    print('>> 번역시작')
    for req_lyric in lyric_list:
        print(req_lyric)
        #request form data 값 선언
        req_data ={
            "source":"en",
            "target":"ko",
            "text":req_lyric
        }
        #papago 서비스 요청, post() 함수 호출
        res= requests.post(url, data=req_data, headers=req_header)
        print(res.status_code)
        try:
            trans_lyric = res.json()['message']['result']['translatedText']
        except Exception as exp:
            print(exp)
        else:
            print(trans_lyric)
            
        trans_list.append(req_lyric+'\n')
        trans_list.append(trans_lyric+'\n')
        
    #print(trans_list)
    translyric_savefiles(trans_list)
    print('>> 번역종료')
main()
```

```python
import requests

#가사 번역 내용 저장 파일
def translyric_savefiles(trans_list, song_name):  
    full_file_name = f'data/{song_name}_trans.txt'
    with open(full_file_name,'w',encoding='utf8')as file:
        file.writelines(trans_list)

#dearname.txt 가사 파일의 내용 읽어서 list에 저장
def gettext_savelist(song_name):
    full_song_name = f'data/{song_name}.txt'
    lyric_list =[]
    with open(full_song_name,'r',encoding='utf8') as file:
        contents = file.read()
        lyric_list = contents.split('\n')
    return lyric_list

def main(song_name):
    client_id = "KadYYszcWbbF7ypJoRpu"
    client_secret = "qEPp5fNh4n"
    
    url = "https://openapi.naver.com/v1/papago/n2mt"

    #request header 값 선언
    req_header = {
        "X-Naver-Client-Id":client_id,
        "X-Naver-Client-Secret":client_secret
    }
    
    #dearname.txt 가사 파일의 내용 읽어서 list에 저장
    lyric_list = gettext_savelist(song_name)
    print(len(lyric_list))
    lyric_list = [lyric for lyric in lyric_list if len(lyric) != 0]
    print(len(lyric_list))
    
    trans_list = []
    print('>> 번역시작')
    for req_lyric in lyric_list:
        print(req_lyric)
        #request form data 값 선언
        req_data ={
            "source":"ko",
            "target":"en",
            "text":req_lyric
        }
        #papago 서비스 요청, post() 함수 호출
        res= requests.post(url, data=req_data, headers=req_header)
        
        try:
            trans_lyric = res.json()['message']['result']['translatedText']
        except Exception as exp:
            print(res.status_code)
        else:
            print(trans_lyric)
            
        trans_list.append(req_lyric+'\n')
        trans_list.append(trans_lyric+'\n')
        
    #print(trans_list)
    translyric_savefiles(trans_list,song_name)
    print('>> 번역종료')
    
main("dearname")
```
